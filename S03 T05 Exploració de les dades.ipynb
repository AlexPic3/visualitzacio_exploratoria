#### Troba quantes dades faltants hi ha per columna

```python
air_nan_df = pd.DataFrame(air_redu_df.isna().sum())
# Percentage of NaN by Column of the total of NaN
air_nan_df["percent"] = round(((air_redu_df.isna().sum())*100)/(air_nan_df.sum()[0]),2)
air_nan_df
```

```python
# Percentage of NaN by Column of the total of NaN
#sns.barplot(x='group', y='Values', data=df, estimator=lambda x: sum(x==0)*100.0/len(x))
plt.figure(figsize=(16, 6))
ax = sns.barplot(x = air_nan_df.index, y= air_nan_df.percent, data= air_nan_df)
ax.set_ylim(0,35)
plt.xticks(color = 'black',rotation = 45, horizontalalignment = 'right')
plt.show()
```

#### Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
import numpy as np
# Velocitat Mitjana Miles/min
air_redu_nc_df = round(air_redu_df.assign(Mean_Speed = (air_redu_df['Distance'] / air_redu_df['AirTime'])*60),2)
# YES OR NO DELAY
air_redu_nc_df['DelayYN'] = np.where(air_redu_df['ArrDelay']>0, 1,0)
air_redu_nc_df = pd.DataFrame(air_redu_nc_df)
#print (air_redu_nc_df)
#print (type(air_redu_nc_df))
air_redu_nc_df
```

```python
# Plot Pie Delayed Or not Delayed
delay_df = air_redu_nc_df.groupby(by= 'DelayYN').count()
delay_df.head()
```

```python
from IPython.core.display import HTML
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")
```

```python
plt.figure(figsize=(12, 6))
fig1, ax1 = plt.subplots()
labels = 'On time', 'Delayed'
explode = (0, 0.1)
ax1.pie(delay_df.UniqueCarrier, explode= explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
fig1.suptitle('Percentage On Time or Delayed Flights', fontsize=20)

plt.show()
```

```python
# Create a new colum with de format of a Timestamp Data using the columns of Year Month and DayofMonth 
air_redu_nc_df["Date"] = pd.to_datetime(dict(year=air_df.Year, month=air_df.Month, day=air_df.DayofMonth))
air_redu_nc_df.head(5)
```

```python
# Plot speed average
plt.figure(figsize=(20, 8))
air_redu_nc_df.groupby(by="Date")["Mean_Speed"].mean().plot(marker='+')
plt.title("Average speed")
plt.ylabel('Miles/h')
plt.grid(True)
plt.show()
```

```python
df = air_redu_nc_df[['UniqueCarrier','Date','Mean_Speed']].groupby(['UniqueCarrier','Date']).mean("Mean_Speed")
#pd.DataFrame(df).reset_index()
df = df.reset_index()
df
```

```python
df.isnull().sum()
```

```python
# Display rows with one or more NaN values in pandas dataframe
df1 = df[df.isna().any(axis=1)]
df1
```

```python
df2 = df.copy()
df2.dropna(subset=['Mean_Speed'],inplace=True)
```

```python
df2.isnull().sum()
```

```python
df1
```

```python
sns.set(rc={'figure.figsize':(24,12)})
sns.set(style="darkgrid")
sns.lineplot(
    data=df2,
    x="Date", y="Mean_Speed", hue="UniqueCarrier")

# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
#uc = df['UniqueCarrier'].unique().tolist()
```

```python
sns.set()
fig, axes = plt.subplots(10, 2, constrained_layout=True, figsize=(23,30), sharex=True)
fig.suptitle('Average Velocity by Air Company', fontsize=25)
ll = ['9E','AA','AQ','AS','B6','CO','DL','EV','F9','FL','HA','MQ','NW','OH','OO','UA','US','WN','XE','YV']

indx = 0
for xrows in range(0,10):
    for ycolum in range(0,2):
        uc = ll[indx]
        df_mask = df['UniqueCarrier'] == uc
        df_i = df[df_mask]
        df_i = df_i.reset_index(drop=True)
        sns.lineplot(ax=axes[xrows, ycolum],x=df_i['Date'], y=df_i['Mean_Speed'])
        axes[xrows, ycolum].set_title(f'Air Company {uc}', size=20)
        indx += 1

#fig.subplots_adjust(wspace=.05, hspace=.3)
plt.show()
```

* **As it is possible to see, there is some difference in the average speed by company, and a futher investigation would be needed to find out which is the couse of this variation. Distance ranges should be taken into account for future reviews.**

#### Taula de les aerolínies amb més endarreriments acumulats

```python
# Calcul del numero de files que conte cada Companyia Aerea
df=pd.DataFrame(air_redu_nc_df["UniqueCarrier"].value_counts())
df.T
#print(type(df))
```

```python
delay_carrier_df = (air_redu_nc_df.groupby(by = ["UniqueCarrier"]).sum()['ArrDelay'])/60
delay_carrier_df = pd.DataFrame(delay_carrier_df.sort_values(axis=0,ascending=False))
delay_carrier5_df = pd.DataFrame(delay_carrier_df[:5])
#print(type(delay_carrier5_df))
delay_carrier5_df.name = 'top 5'
print("Les 5 primeras companyies amb una mayor acumulació de retards en HORES\nen el an 2008 són: \n ")
round(delay_carrier5_df,1)
```

```python
delay_carrier_df.info()
```

```python
ax = sns.barplot(x=delay_carrier_df.index, y=delay_carrier_df.ArrDelay, data=delay_carrier_df)
```

#### Quins són els vols més llargs? 

```python
#Els vols mes llargs són:
air_redu_nc_UFD = air_redu_nc_df[['UniqueCarrier','FlightNum','Distance']]
# print (air_redu_nc_UFD.sort_values(by=['Distance'], axis=0, ascending=False))
air_redu_nc_mll = air_redu_nc_UFD.groupby(['UniqueCarrier','Distance']).count().sort_values(by=['Distance'],ascending=False).head(6)
air_redu_nc_mll
```

```python
air_redu_nc_mll.reset_index(inplace=True)
```

```python
air_redu_nc_mll.info()
```

```python
ax = sns.barplot(x=air_redu_nc_mll.UniqueCarrier, y=air_redu_nc_mll.Distance, data=air_redu_nc_mll)
```

#### I els més endarrerits?

```python
#Els vols mes enraderits són:
air_redu_nc_UFA = air_redu_nc_df[['UniqueCarrier','FlightNum','ArrDelay']]
#print (air_redu_nc_UFA.sort_values(by=['ArrDelay'], axis=0, ascending=False))
air_redu_nc_Mde = air_redu_nc_UFA.groupby(['UniqueCarrier','ArrDelay']).count().sort_values(by=['ArrDelay'],ascending=False).head(9)
air_redu_nc_Mde

```

```python
air_redu_nc_Mde.reset_index(inplace=True)
```

```python
ax = sns.barplot(x=air_redu_nc_Mde.UniqueCarrier, y=air_redu_nc_Mde.ArrDelay, data=air_redu_nc_Mde)
```

* **The 3 worst delays belong to the same Air company NW = Northwest Airlines. This is why it shows only three bars**
